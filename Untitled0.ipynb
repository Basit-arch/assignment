 Qno.6 How do conditional statements work in Python? Illustrate with examples

 Conditional statements in Python control program flow based on conditions using if, elif, and
 else. The code block runs if the condition is True; otherwise, it moves to the next condition or
 exits.
 Example


```python
x = 10
if x > 5:
    print("Greater")
elif x == 5:
    print("Equal")
else:
    print("Smaller")
```

 Qno.7 Describe the different types of loops in Python and their use cases with
 examples

 Python has two main loop types:
  for loops and while loops.

 •For loops iterate over a sequence (like a list, tuple, or string) and are useful for known,
 finite iterations.

```python
for i in range(5):
 print(i) # Prints 0 to 4
```

•while loops run as long as a condition is true, ideal for unknown or conditional iterations

```python
x = 0
while x < 5:
  print(x)
x += 1
 # Prints 0 to 4
```

 These loops help automate repetitive tasks
